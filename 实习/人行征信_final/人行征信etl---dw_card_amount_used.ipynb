```python
# 需要注意group by t1.reportno 这一段

spark.sql("""

select 
	head.reportno,
	case when ccard.reportno is null then -9999999 
		else nvl(ccardused_over50_account_num,0) 
	end as ccardused_over50_account_num,
	case when ccard.reportno is null then -9999999 
		else nvl(ccardused_over75_account_num,0) 
	end as ccardused_over75_account_num,	
	case when ccard.reportno is null then -9999999 
		else nvl(ccardused_over90_account_num,0) 
	end as ccardused_over90_account_num,
	case when scard.reportno is null then -9999999 
		else nvl(scardused_over50_account_num,0) 
	end as scardused_over50_account_num,
	case when scard.reportno is null then -9999999 
		else nvl(scardused_over75_account_num,0) 
	end as scardused_over75_account_num,
	case when scard.reportno is null then -9999999 
		else nvl(scardused_over90_account_num,0) 
	end as scardused_over90_account_num,
	case when ccard.reportno is null then -9999999 
		when ccard.reportno is not null and ccardused_over50_account_percent=-9999998 then -9999998
		else nvl(ccardused_over50_account_percent,0) 
	end as ccardused_over50_account_percent,
	case when ccard.reportno is null then -9999999 
		when ccard.reportno is not null and ccardused_over75_account_percent=-9999998 then -9999998
		else nvl(ccardused_over75_account_percent,0)  
	end as ccardused_over75_account_percent,		
	case when ccard.reportno is null then -9999999 
		when ccard.reportno is not null and ccardused_over90_account_percent=-9999998 then -9999998
		else nvl(ccardused_over90_account_percent,0)  
	end as ccardused_over90_account_percent,		
	case when scard.reportno is null then -9999999 
		when scard.reportno is not null and scardused_over50_account_percent=-9999998 then -9999998
		else nvl(scardused_over50_account_percent,0)  
	end as scardused_over50_account_percent,		
	case when scard.reportno is null then -9999999 
		when scard.reportno is not null and scardused_over75_account_percent=-9999998 then -9999998
		else nvl(scardused_over75_account_percent,0)  
	end as scardused_over75_account_percent,
	case when scard.reportno is null then -9999999 
		when scard.reportno is not null and scardused_over90_account_percent=-9999998 then -9999998
		else nvl(scardused_over90_account_percent,0) 
	end as scardused_over90_account_percent,
	now() as create_time,now() as update_time,now() as etl_date


from ICRMESSAGEHEADER head

left join 
(
	select 
		t1.reportno,
		nvl(count(distinct ccardused_over50_account_num),0) as ccardused_over50_account_num,
		nvl(count(distinct ccardused_over75_account_num),0) as ccardused_over75_account_num,
		nvl(count(distinct ccardused_over90_account_num),0) as ccardused_over90_account_num,
		nvl(count(distinct scardused_over50_account_num),0) as scardused_over50_account_num,
		nvl(count(distinct scardused_over75_account_num),0) as scardused_over75_account_num,
		nvl(count(distinct scardused_over90_account_num),0) as scardused_over90_account_num,
		case when nvl(t2.ACCOUNTCOUNT,0) =0 then -9999998 else round(count(distinct ccardused_over50_account_num)/t2.ACCOUNTCOUNT,5) end as ccardused_over50_account_percent,
		case when nvl(t2.ACCOUNTCOUNT,0) =0 then -9999998 else round(count(distinct ccardused_over75_account_num)/t2.ACCOUNTCOUNT,5) end as ccardused_over75_account_percent,
		case when nvl(t2.ACCOUNTCOUNT,0) =0 then -9999998 else round(count(distinct ccardused_over90_account_num)/t2.ACCOUNTCOUNT,5) end as ccardused_over90_account_percent,
		case when nvl(t3.ACCOUNTCOUNT,0) =0 then -9999998 else round(count(distinct scardused_over50_account_num)/t3.ACCOUNTCOUNT,5) end as scardused_over50_account_percent,
		case when nvl(t3.ACCOUNTCOUNT,0) =0 then -9999998 else round(count(distinct scardused_over75_account_num)/t3.ACCOUNTCOUNT,5) end as scardused_over75_account_percent,
		case when nvl(t3.ACCOUNTCOUNT,0) =0 then -9999998 else round(count(distinct scardused_over90_account_num)/t3.ACCOUNTCOUNT,5) end as scardused_over90_account_percent
		
	from (
	select 
			reportno,
			case when BIZTYPE='贷记卡' and used_percent>0.5 then account else null end as ccardused_over50_account_num,
			case when BIZTYPE='贷记卡' and used_percent>0.75 then account else null end as ccardused_over75_account_num,
			case when BIZTYPE='贷记卡' and used_percent>0.9 then account else null end as ccardused_over90_account_num,
			case when BIZTYPE='准贷记卡' and used_percent>0.5 then account else null end as scardused_over50_account_num,
			case when BIZTYPE='准贷记卡' and used_percent>0.75 then account else null end as scardused_over75_account_num,
			case when BIZTYPE='准贷记卡' and used_percent>0.9 then account else null end as scardused_over90_account_num
	
	from (
		select 
			h.reportno,
			t.account,
			t.BIZTYPE,
			t.used_percent
			from ICRMESSAGEHEADER h
		left join 
			(
			select 
					reportno,
					account,
					BIZTYPE,
					nvl(USEDCREDITLIMITAMOUNT,0)/CREDITLIMITAMOUNT as used_percent
			from ICRLOANCARDINFO
			where CREDITLIMITAMOUNT!=0
			) t on h.reportno=t.reportno
		)t
	)t1
		left join (select reportno,ACCOUNTCOUNT from ICRUNDESTORYLOANCARD)t2 on t1.reportno=t2.reportno
		left join (select reportno,ACCOUNTCOUNT from ICRUNDESTORYSTANDARDLOANCARD) t3 on t1.reportno=t3.reportno
	group by t1.reportno,t2.ACCOUNTCOUNT,t3.ACCOUNTCOUNT
)main on head.reportno=main.reportno
left join
(
	select 
		  distinct reportno 
	from ICRLOANCARDINFO 
	where biztype='贷记卡'
)ccard on head.reportno=ccard.reportno
left join
(
	select 
		distinct reportno 
	from ICRLOANCARDINFO 
	where biztype='准贷记卡'
)scard on head.reportno=scard.reportno

""").createOrReplaceTempView("dw_card_amount_used")
```
