```python
spark.sql("""

select 
	distinct REPORTNO,
	account,
	month,
	gap_months,
	row_number() over (partition by REPORTNO order by REPORTNO,month,account ) as rownum

from temp_loan_5yearoverdue
where gap_months<24
order by REPORTNO,month,account

""").createOrReplaceTempView("temp_loan_cont_1")

```

```python
spark.sql("""

select 
	reportno,
	account,
	month,
	gap_months,
	rownum,
 	add_months(date_format(concat(month,'-01'),'%Y-%m-%d'),  -1*rownum ) as timeadd
from temp_loan_cont_1

""").createOrReplaceTempView("temp_loan_cont_2")
```

```python
spark.sql("""


select head.reportno,
	case when loan.reportno is null then -9999999
		else nvl(m3.mth3_loan_overdue_maxmonth,0) end as mth3_loan_overdue_maxmonth,
	case when loan.reportno is null then -9999999
		else nvl(m6.mth6_loan_overdue_maxmonth,0) end as mth6_loan_overdue_maxmonth,
	case when loan.reportno is null then -9999999
		else nvl(m9.mth9_loan_overdue_maxmonth,0) end as mth9_loan_overdue_maxmonth,
	case when loan.reportno is null then -9999999
		else nvl(m12.mth12_loan_overdue_maxmonth,0) end as mth12_loan_overdue_maxmonth,
	case when loan.reportno is null then -9999999
		else nvl(m18.mth18_loan_overdue_maxmonth,0) end as mth18_loan_overdue_maxmonth,
	case when loan.reportno is null then -9999999
		else nvl(m24.mth24_loan_overdue_maxmonth,0) end as mth24_loan_overdue_maxmonth,
	now() as create_time,now() as update_time,now() as etl_date
from ICRMESSAGEHEADER head
left join 
(select reportno,max(cont_mons) as mth3_loan_overdue_maxmonth
from (
	select reportno,account,count(*) as cont_mons
	from temp_loan_cont_2
	where gap_months<3
	group by reportno,account,timeadd
	)t
group by reportno	
)m3 on head.reportno=m3.reportno
left join 
(select reportno,max(cont_mons) as mth6_loan_overdue_maxmonth
from (
	select reportno,account,count(*) as cont_mons
	from temp_loan_cont_2
	where gap_months<6
	group by reportno,account,timeadd
	)t
group by reportno	
)m6 on head.reportno=m6.reportno
left join 
(select reportno,max(cont_mons) as mth9_loan_overdue_maxmonth
from (
	select reportno,account,count(*) as cont_mons
	from temp_loan_cont_2
	where gap_months<9
	group by reportno,account,timeadd
	)t
group by reportno	
)m9 on head.reportno=m9.reportno
left join 
(select reportno,max(cont_mons) as mth12_loan_overdue_maxmonth
from (
	select reportno,account,count(*) as cont_mons
	from temp_loan_cont_2
	where gap_months<12
	group by reportno,account,timeadd
	)t
group by reportno	
)m12 on head.reportno=m12.reportno
left join 
(select reportno,max(cont_mons) as mth18_loan_overdue_maxmonth
from (
	select reportno,account,count(*) as cont_mons
	from temp_loan_cont_2
	where gap_months<18
	group by reportno,account,timeadd
	)t
group by reportno	
)m18 on head.reportno=m18.reportno
left join 
(select reportno,max(cont_mons) as mth24_loan_overdue_maxmonth
from (
	select reportno,account,count(*) as cont_mons
	from temp_loan_cont_2
	where gap_months<24
	group by reportno,account,timeadd
	)t
group by reportno	
)m24 on head.reportno=m24.reportno
left join 
(select distinct reportno from ICRLOANINFO
)loan  on head.reportno=loan.reportno


""").createOrReplaceTempView("dw_loan_contiue_overdue")
```
