```python
spark.sql("""



select 
		reportno,
		sum(latest1mon_new_loan_num) as latest1mon_new_loan_num,
		sum(latest3mon_new_loan_num) as latest3mon_new_loan_num,
		sum(latest6mon_new_loan_num) as latest6mon_new_loan_num,
		sum(latest12mon_new_loan_num) as latest12mon_new_loan_num,
		sum(latest1mon_new_loan_amount) as latest1mon_new_loan_amount,
		sum(latest3mon_new_loan_amount) as latest3mon_new_loan_amount,
		sum(latest6mon_new_loan_amount) as latest6mon_new_loan_amount,
		sum(latest12mon_new_loan_amount) as latest12mon_new_loan_amount

from (
		select 
				reportno,
				case when gap_days<=30 then 1 else 0 end as latest1mon_new_loan_num,
				case when gap_days<=90 then 1 else 0 end as latest3mon_new_loan_num,
				case when gap_days<=180 then 1 else 0 end as latest6mon_new_loan_num,
				case when gap_days<=360 then 1 else 0 end as latest12mon_new_loan_num,
	
				case when gap_days<=30 then nvl(CREDITLIMITAMOUNT,0) else 0 end as latest1mon_new_loan_amount,
				case when gap_days<=90 then nvl(CREDITLIMITAMOUNT,0) else 0 end as latest3mon_new_loan_amount,
				case when gap_days<=180 then nvl(CREDITLIMITAMOUNT,0) else 0 end as latest6mon_new_loan_amount,
				case when gap_days<=360 then nvl(CREDITLIMITAMOUNT,0) else 0 end as latest12mon_new_loan_amount

	from 
	(

			select 
					lofo.reportno,
					lofo.CREDITLIMITAMOUNT,
					day(date_format(head.QUERYTIME,'%Y-%m-%d'))-day(date_format(lofo.OPENDATE,'%Y-%m-%d')) as gap_days 
		from ICRLOANINFO lofo
		left join ICRMESSAGEHEADER  head on lofo.reportno=head.reportno
		-- where TIMESTAMPDIFF(day,date_format(head.QUERYTIME,'%Y-%m-%d'),date_format(lofo.OPENDATE,'%Y-%m-%d'))<=360
	)t
		WHERE gap_days<=360
) t1
group by reportno


""").createOrReplaceTempView("dw_new_loan")
```

```python
sql = """

select reportno,
sum(latest1mon_new_loan_num) as latest1mon_new_loan_num,
sum(latest3mon_new_loan_num) as latest3mon_new_loan_num,
sum(latest6mon_new_loan_num) as latest6mon_new_loan_num,
sum(latest12mon_new_loan_num) as latest12mon_new_loan_num,
sum(latest1mon_new_loan_amount) as latest1mon_new_loan_amount,
sum(latest3mon_new_loan_amount) as latest3mon_new_loan_amount,
sum(latest6mon_new_loan_amount) as latest6mon_new_loan_amount,
sum(latest12mon_new_loan_amount) as latest12mon_new_loan_amount

from (
	select reportno,
	case when gap_days<=30 then 1 else 0 end as latest1mon_new_loan_num,
	case when gap_days<=90 then 1 else 0 end as latest3mon_new_loan_num,
	case when gap_days<=180 then 1 else 0 end as latest6mon_new_loan_num,
	case when gap_days<=360 then 1 else 0 end as latest12mon_new_loan_num,
	
	case when gap_days<=30 then ifnull(CREDITLIMITAMOUNT,0) else 0 end as latest1mon_new_loan_amount,
	case when gap_days<=90 then ifnull(CREDITLIMITAMOUNT,0) else 0 end as latest3mon_new_loan_amount,
	case when gap_days<=180 then ifnull(CREDITLIMITAMOUNT,0) else 0 end as latest6mon_new_loan_amount,
	case when gap_days<=360 then ifnull(CREDITLIMITAMOUNT,0) else 0 end as latest12mon_new_loan_amount

	from 
	(

		select lofo.reportno,lofo.CREDITLIMITAMOUNT,
			TIMESTAMPDIFF(day,date_format(lofo.OPENDATE,'%%Y-%%m-%%d'),date_format(head.QUERYTIME,'%%Y-%%m-%%d')) as gap_days 
		from ICR_PROD.ICRLOANINFO lofo
		left join ICR_PROD.ICRMESSAGEHEADER  head on lofo.reportno=head.reportno
		-- where TIMESTAMPDIFF(day,date_format(head.QUERYTIME,'%%Y-%%m-%%d'),date_format(lofo.OPENDATE,'%%Y-%%m-%%d'))<=360
	)t
		WHERE gap_days<=360
) t1
group by reportno;

"""
```
